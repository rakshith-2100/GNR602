```python
def calculate_glcm(image, dx, dy):
    upper_range=np.amax(a[0])
    # Define the GLCM matrix with zeros
    glcm = np.zeros((upper_range+1, upper_range+1), dtype=np.uint32)
    rows, cols = image.shape
    for i in range(rows):
        for j in range(cols):
            # Get the current pixel value
            current_pixel = image[i, j]
            neighbor_i = i + dy
            neighbor_j = j + dx

            # Check if the neighbor pixel is within the image bounds
            if 0 <= neighbor_i < rows and 0 <= neighbor_j < cols:
                neighbor_pixel = image[neighbor_i, neighbor_j]
                glcm[current_pixel, neighbor_pixel] += 1

    return glcm
```

```python
GLCM_matrix=calculate_glcm(a[0],dx=1,dy=0)
```

```python
eigen_val=np.real(np.linalg.eigvals(GLCM_matrix))
```

```python
window_size = 40
rows, cols = gt_data.shape
X_train=[]
Y_train=[]
for i in trange(rows - window_size + 1):
    for j in range(cols - window_size + 1):
        window = input_data[i:i+window_size,j:j+window_size]
        gt_window = gt_data[i:i+window_size, j:j+window_size]
        unique, counts = np.unique(gt_window, return_counts=True)
        most_common_value = unique[np.argmax(counts)]
        GLCM_matrix=calculate_glcm(window,dx=1,dy=0)
        X_train+=[np.real(np.linalg.eigvals(GLCM_matrix))]
        Y_train+=[most_common_value]
```
